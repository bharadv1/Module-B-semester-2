# 2. Weekly graph question

Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='ME')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='ME')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

One way to get the viewer to focus on the last three months first would be to use preattentive attributes to visually emphasize those points. For example, the first nine months could be shown in a light gray while the final three months are highlighted in a darker or brighter color. This contrast would naturally draw the viewer’s attention to the most recent data. Another option would be to increase the marker size or line thickness for the last three points, or add labels to those values. By reducing the visual emphasis on the earlier months and highlighting the final months, the viewer is more likely to immediately examine the most recent part of the chart.

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt

months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

received = [160,185,240,150,180,160,130,200,160,140,150,175]
processed = [160,185,235,148,180,150,123,155,125,102,125,140]

plt.figure(figsize=(9,5))

plt.plot(months, received, color="#4C78A8", linewidth=3)
plt.plot(months, processed, color="#C44E52", linewidth=3)

plt.ylim(0,300)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.text(11.3, received[-1], "Received", color="#4C78A8", fontsize=16, va='center')
plt.text(11.3, processed[-1], "Processed", color="#C44E52", fontsize=16, va='center')

plt.tight_layout()
plt.show()
```
