# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

The first approach, where the two weeks are plotted in separate graphs, makes each individual time series easier to read. There is no overlap between the lines, so it is clear how the number of visitors changes throughout each week. This is useful if the goal is to understand the pattern within each week independently. However, the main drawback is that it makes comparison between the two weeks more difficult, since the viewer has to mentally switch between the two plots and estimate differences. The second approach, where both weeks are plotted on the same graph, makes comparison much easier. By placing both lines on the same axes, it becomes immediately clear how the number of visitors differs across the same days in each week. This is especially useful for identifying trends or shifts between weeks. However, the downside is that the graph can become more cluttered, and overlapping lines may reduce readability, especially if more series were added. Overall, the combined plot is more effective when the goal is comparison, while the separate plots are better when the goal is to clearly understand each time series on its own.

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 187-205 of the Storytelling With Data book as best you can. (Chapter Eight). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt
import numpy as np

products = ["Product A", "Product B", "Product C", "Product D", "Product E"]
years = ["2008", "2009", "2010", "2011", "2012", "2013", "2014"]

data = {
    "2008": [420, 410, 100, 150, 0],
    "2009": [430, 420, 200, 180, 0],
    "2010": [400, 390, 180, 200, 0],
    "2011": [390, 300, 250, 260, 0],
    "2012": [280, 260, 180, 220, 100],
    "2013": [260, 250, 170, 210, 0],
    "2014": [270, 260, 190, 220, 210],
}

x = np.arange(len(products))
width = 0.1

plt.figure(figsize=(10,6))

colors = ["#d9d9d9", "#bdbdbd", "#969696", "#737373", "#525252", "#252525", "#000000"]

for i, year in enumerate(years):
    plt.bar(x + i*width, data[year], width, color=colors[i], label=year)

plt.xticks(x + width*3, products)
plt.ylabel("Price ($)")
plt.title("Average Retail Product Price Per Year")

plt.legend(title="Year", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```
